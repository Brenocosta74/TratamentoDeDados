```python
import pandas as pd

#1. Carregar os dados do arquivo
df = pd.read_excel('dados_ausentes.xlsx')
display(df)

#2. Verificar se há dados ausentes
display(df.isnull())

#3. Contar os valores ausentes
display(df.isnull().sum())

```

```python
#2. Verificar se há dados ausentes
display(df.isnull())
```

```python
#3. Contar os valores ausentes
display(df.isnull().sum())
```

```python
# Excluir as linhas com dados ausentes
df_sem_nulos = df.dropna()
display(df_sem_nulos)
```

```python
# Substituir dados ausentes
# Cidades - Desconhecido

df['Cidade'].fillna('Desconhecida', inplace=True)
display(df)
```

```python
# Substituir dados ausentes
#Substitui 'Salario' por 0
df['Salário'].fillna(0, inplace=True)

display(df)
```

```python
# Substituir dados ausentes
# Substitui 'Idade' pela media

df['Idade'].fillna(df['Idade'].mean(), inplace=True)
display(df)
```
